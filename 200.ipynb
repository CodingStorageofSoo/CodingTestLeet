```python
def bfs (i, j):
        
    q = deque ()
    q.append([i, j])

    while q :

        x, y = q.popleft()
        grid[x][y] = "0"
        

        for nx, ny in ((x-1,y), (x+1,y), (x,y-1), (x,y+1)):

            if 0 <= nx < len(grid) and 0 <= ny < len(grid[0]) and grid[nx][ny] == "1":
                q.append([nx, ny])
       
```

```python
cnt = 0
for i in range(len(grid)):
    for j in range(len(grid[0])):
        if grid[i][j] == "1":
            bfs (i, j)
            cnt += 1

print(cnt)
```
